```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math
# Import CSV file 
gdp_data = pd.read_csv('all_data.csv')
# Inspect the dataframe
print(gdp_data.head())
sns.histplot(gdp_data.GDP)
plt.show()
plt.clf()
```

```python
# Gather unique Countries and save them as countries then print
countries = gdp_data.Country.unique()
print(countries)
# Gather unique years and save them as years then print
years = gdp_data.Year.unique()
print(years)
```

```python
# Save each countries as a seperate variable 
chile = gdp_data[gdp_data.Country == 'Chile']
china = gdp_data[gdp_data.Country == 'China']
germany = gdp_data[gdp_data.Country == 'Germany']
mexico = gdp_data[gdp_data.Country == 'Mexico']
USA = gdp_data[gdp_data.Country == 'United States of America']
zimbabwe = gdp_data[gdp_data.Country == 'Zimbabwe']
# Plot each country as a seperate line on the plot
sns.lineplot(data = chile, x='Year', y='GDP', size_norm =True)
sns.lineplot(data = china, x='Year', y='GDP', size_norm =True)
sns.lineplot(data = germany, x='Year', y='GDP', size_norm =True)
sns.lineplot(data = mexico, x='Year', y='GDP', size_norm =True)
sns.lineplot(data = USA, x='Year', y='GDP', size_norm =True)
sns.lineplot(data = zimbabwe, x='Year', y='GDP', size_norm =True)
plt.legend(countries, loc= 'upper left',  prop={'size': 6.5})
# Plot the title, x label, and y label
plt.title('GDP by Country from 2000 - 2015')
plt.ylabel('GDP (in billions)')
plt.xlabel('Year')
# Show the graph
plt.show()
plt.clf()
```

```python
# Plot lineplot, year vs Life Expectancy at birth
sns.lineplot(data = chile, x='Year', y='Life expectancy at birth (years)', size_norm =True)
sns.lineplot(data = china, x='Year', y='Life expectancy at birth (years)', size_norm =True)
sns.lineplot(data = germany, x='Year', y='Life expectancy at birth (years)', size_norm =True)
sns.lineplot(data = mexico, x='Year', y='Life expectancy at birth (years)', size_norm =True)
sns.lineplot(data = USA, x='Year', y='Life expectancy at birth (years)', size_norm =True)
sns.lineplot(data = zimbabwe, x='Year', y='Life expectancy at birth (years)', size_norm =True)
plt.legend(countries, loc= 'upper left',  prop={'size': 6.5})
# Plot the title, x label, and y label
plt.title('Life expectancy at birth without zimbabwe')
plt.ylabel('GDP (in billions)')
plt.show()
plt.clf()
```
